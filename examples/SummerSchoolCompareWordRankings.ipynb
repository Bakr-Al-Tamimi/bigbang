```python
from bigbang.archive import Archive
from bigbang.archive import load as load_archive
import os
import pandas as pd
import numpy as np
```

```python

icann_path = "../archives/http:/mm.icann.org/pipermail"
ncuc_path = "../archives/http:/lists.ncuc.org/pipermail"

paths = [os.path.join(icann_path,"ipc-gnso.csv"),
        os.path.join(icann_path,"wp4.csv"),
        os.path.join(ncuc_path,"ncuc-discuss.csv")]

archives = [load_archive(path) for path in paths]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

#tp = u'(?u)\x08[^\\W\\d_][^\\W\\d_]+\x08'

tp = u'(?u)\\b[^\\W\\d\_]\\w+\\b'

```

```python
def ordered_words(archive):
    cv = CountVectorizer(max_df=.16,min_df=5,token_pattern=tp)
    
    c_dtm = cv.fit_transform(archive.data['Body'].dropna())
    
    feature_names = cv.get_feature_names()
    feature_counts = np.array(c_dtm.sum(axis=0))
    
    feature_order = np.array(np.argsort(feature_counts))[0][::-1]
    
    sorted_features = [feature_names[i] for i in feature_order]
    
    rankings = pd.Series({pair[1] : pair[0] 
                          for pair 
                          in enumerate(sorted_features)})

    return rankings
```

```python
rankings = pd.concat([ordered_words(archive) for archive in archives],axis=1)
```
